```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load dataset as DataFrame
data = fetch_california_housing(as_frame=True)
df = data['frame']

```

```python
# Inspect the dataset
print(df.info())
print(df.describe())
print(df.head())
```

Problem 2 – Subsetting

```python

#1. Filter houses where MedInc > 8.0, save as high income. Display the result.
high_income = df[df['MedInc'] > 8.0]
print("High Income Houses:")
print(high_income.head())
```

```python
#2. Filter houses where Latitude > 37, save as north california. Display the result.
north_california = df[df['Latitude'] > 37]
print("\nNorth California Houses:")
print(north_california.head())
```

```python
#3. Filter houses where AveRooms > 6.0 and AveOccup < 2.0, save as spacious low occupancy. Display the result.
spacious_low_occupancy = df[(df['AveRooms'] > 6.0) & (df['AveOccup'] < 2.0)]
print("\nSpacious Low Occupancy Houses:")
print(spacious_low_occupancy.head())

```

```python
# Subsetting Categorical Equivalents:

#1. Create a new column Region based on Latitude values:
# • ’North’ if Latitude > 37
# • ’Central’ if 35 < Latitude ≤ 37
# • ’South’ otherwise

def get_region(lat):
    if lat > 37:
        return 'North'
    elif lat > 35:
        return 'Central'
    else:
        return 'South'

df['Region'] = df['Latitude'].apply(get_region)
```

```python
# Filter houses where Region is ’North’ or ’Central’, save as north central region. Display the
# result.

north_central_region = df[df['Region'].isin(['North', 'Central'])]
print("\nNorth + Central Region Houses:")
print(north_central_region.head())
```

Problem – 3 Exploratory Data Analysis:


```python


# Question no 1- Which house has the highest value per room?
df['value_per_room'] = df['MedHouseVal'] / df['AveRooms']
high_vpr = df[df['value_per_room'] > 1]
high_vpr_sorted = high_vpr.sort_values(by='value_per_room', ascending=False)

print("Top 5 Houses with Highest Value per Room:")
print(high_vpr_sorted[['MedHouseVal', 'AveRooms', 'value_per_room']].head())
```

```python
# Question n0 2 - Among high-population areas (Population > 5000), which have the highest median income per
# person?

df['income_per_person'] = df['MedInc'] / df['Population']
dense_areas = df[df['Population'] > 5000]
rich_dense_areas = dense_areas.sort_values(by='income_per_person', ascending=False)

print("\nTop 5 Rich Dense Areas (Income per Person):")
print(rich_dense_areas[['MedInc', 'Population', 'income_per_person']].head())
```

Problem – 4 Group By Exercises:


```python
# Question no 1 -  What percent of total house value comes from each Region?

total_val = df['MedHouseVal'].sum()
region_val = df.groupby('Region')['MedHouseVal'].sum()
region_percent = (region_val / total_val) * 100
print("Percent of Total House Value by Region:")
print(region_percent)
```

```python
# Question no 2 - What percent of total houses belong to different age groups?

def age_group(age):
    if age < 20:
        return 'New'
    elif age < 40:
        return 'Mid'
    else:
        return 'Old'

df['AgeGroup'] = df['HouseAge'].apply(age_group)
group_counts = df['AgeGroup'].value_counts()
age_group_percent = (group_counts / len(df)) * 100
print("\nPercent of Houses by Age Group:")
print(age_group_percent)
```

```python
# 2. Handling Missing Data:
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#• Randomly set 5% of AveRooms values to NaN (simulate missingness).
df_missing = df.copy()
num_missing = int(0.05 * len(df_missing))
missing_idx = np.random.choice(df_missing.index, num_missing, replace=False)
df_missing.loc[missing_idx, 'AveRooms'] = np.nan
```

```python
#• Impute missing values using median imputation.
median_val = df_missing['AveRooms'].median()                     
df_missing['AveRooms'] = df_missing['AveRooms'].fillna(median_val)
```

```python
# • Visualize and compare distributions of AveRooms before and after imputation using histograms or
# boxplots.

# Histogram Comparison
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['AveRooms'], kde=True, color='blue', label='Original')
sns.histplot(df_missing['AveRooms'], kde=True, color='green', label='After Imputation', alpha=0.5)
plt.title('Histogram of AveRooms')
plt.xlabel('AveRooms')
plt.legend()

# Boxplot Comparision
boxplot_df = pd.DataFrame({
    'AveRooms': pd.concat([df['AveRooms'], df_missing['AveRooms']]),
    'Group': ['Original'] * len(df) + ['After Imputation'] * len(df_missing)
})

plt.subplot(1, 2, 2)
sns.boxplot(x='Group', y='AveRooms', data=boxplot_df, 
            palette={'Original': 'blue', 'After Imputation': 'green'})
plt.title('Boxplot of AveRooms')

plt.tight_layout()
plt.show()
```

```python
# 2. Vectorization vs Loops
# • Write a function to compute element-wise square of an array using:
# – a for-loop
# – NumPy vectorized operation
# • Compare their execution time using %%timeit or time module.
def square_with_loop(arr):
    result = np.empty_like(arr)
    for i in range(arr.shape[0]):
        for j in range(arr.shape[1]):
            result[i, j] = arr[i, j] ** 2
    return result

def square_vectorized(arr):
    return arr ** 2
```

```python
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
data = fetch_california_housing(as_frame=True)
df = data.frame

# 1. First create the Region column based on Latitude
df['Region'] = pd.cut(df['Latitude'],
                     bins=[-float('inf'), 35, 37, float('inf')],
                     labels=['South', 'Central', 'North'])

# 1. Plot histogram of MedHouseVal
plt.figure(figsize=(10, 6))
plt.hist(df['MedHouseVal'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Median House Values in California', fontsize=14)
plt.xlabel('Median House Value ($100,000s)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.show()

# 2. Now create the bar chart (Region column exists now)
region_income = df.groupby('Region')['MedInc'].mean().sort_values()

plt.figure(figsize=(8, 5))
bars = plt.bar(region_income.index, region_income.values,
              color=['#4c72b0', '#55a868', '#c44e52'])

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.2f}',
            ha='center', va='bottom')

plt.title('Average Median Income by California Region', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Average Median Income', fontsize=12)
plt.ylim(0, region_income.max() * 1.1)
plt.show()
```

```python
# Problem 3 – Scatter Plots
import seaborn as sns
# 1. Create a scatter plot of MedInc vs. MedHouseVal.
plt.figure(figsize=(10, 6))
plt.scatter(df['MedInc'], df['MedHouseVal'])
plt.title('Median Income vs. House Value')
plt.xlabel('Median Income (in $10,000s)')
plt.ylabel('Median House Value (in $100,000s)')
plt.grid(True)
plt.show()

# 2. Color the points by Region and add transparency.
df['Region'] = pd.cut(df['Latitude'],
                     bins=[-float('inf'), 35, 37, float('inf')],
                     labels=['South', 'Central', 'North'])

# Scatter plot with Region colors and transparency
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['MedInc'], df['MedHouseVal'],
                     c=df['Region'].cat.codes,  
                     cmap='viridis',             
                     alpha=0.5,                  
                     s=50)                       


plt.title('Median Income vs. House Value by Region')
plt.xlabel('Median Income (in $10,000s)')
plt.ylabel('Median House Value (in $100,000s)')
plt.grid(True)
plt.show()

# 3. Add a regression line using Seaborn’s regplot.
sns.regplot(
    data=df,
    x='MedInc',
    y='MedHouseVal',
    scatter=False,  
    color='red',    
    line_kws={'linewidth': 2.5}  
)
```

```python
# Problem 4 – Subplots
# 1. Create a 2x2 subplot grid showing:
# • Line plot of sine
# • Histogram of income
# • Bar chart of region-wise population
# • Boxplot of house value grouped by age group

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load data and prepare features
data = fetch_california_housing(as_frame=True)
df = data.frame

# Create required categorical columns FIRST
df['Region'] = pd.cut(df['Latitude'], 
                     bins=[-float('inf'), 35, 37, float('inf')],
                     labels=['South', 'Central', 'North'])

df['AgeGroup'] = pd.cut(df['HouseAge'],
                       bins=[0, 20, 40, float('inf')],
                       labels=['New', 'Mid', 'Old'])

# Create figure and subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# 1. Line plot of sine function
x = np.linspace(0, 2*np.pi, 100)
axes[0,0].plot(x, np.sin(x), color='blue')
axes[0,0].set_title('Sine Function')
axes[0,0].grid(True)

# 2. Histogram of income
axes[0,1].hist(df['MedInc'], bins=30, color='green', alpha=0.7)
axes[0,1].set_title('Income Distribution')
axes[0,1].set_xlabel('Median Income')

# 3. Bar chart of region-wise population
region_pop = df.groupby('Region')['Population'].sum()
axes[1,0].bar(region_pop.index, region_pop.values, color=['red', 'blue', 'green'])
axes[1,0].set_title('Population by Region')
axes[1,0].set_ylabel('Total Population')

# 4. Boxplot of house value by age group (NOW CORRECT)
sns.boxplot(data=df, x='AgeGroup', y='MedHouseVal', ax=axes[1,1], palette='pastel')
axes[1,1].set_title('House Value by Age Group')
axes[1,1].set_ylabel('Median House Value')

plt.tight_layout()
plt.show()
```
